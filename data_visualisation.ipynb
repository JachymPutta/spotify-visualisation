# Visualising the Million Playlist Dataset

# Introduction
The Million Playlist dataset (MPD) is the largest public dataset of music playlists in the world. It has been created by sampling over 2 billion playlists.
The entire playlist has been anonymised to perserve user anonymity. Data visualisation should be one of the first steps when dealing with a new dataset.
It allows us to extract information, interpretable by humans. A good data visualisation can extract underlying structures and generate novel insights 
which might have gone under the radar before, without visualisation raw data stays uninterpretable from humans, hiding a wealth of information.

The following notebook will explore the Million Playlist Dataset in two main ways. We will begin by visualising the raw data without modification. This
will allow us to extract simple trends like the most popular songs and artist. The second way requires some modifications to the underlying data, to enable
more complex techniques like *Principal Component Analysis* or *K-means clustering*.

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import concurrent.futures
import glob
import json
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from dotenv import load_dotenv
import os
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from wordcloud import WordCloud
from sklearn.cluster import KMeans

if not load_dotenv():
    print("No .env file found, using default environment variables")
else:
    print("Using environment variables from .env file")
````

## Loading data

Let's start by fetching the playlist data from the `RAW_FILES` folder.

````python
raw_files = os.getenv("RAW_FILES", "docs/sample_data/mpd.slice.*.json")

def load_file(filename):
    with open(filename) as f:
        json_data = json.load(f)
        for playlist in json_data["playlists"]:
            for track in playlist["tracks"]:
                track_uris.append(track["track_uri"].split("spotify:track:")[1])
        data_frame = pd.DataFrame(json_data['playlists']) 
        return data_frame

filenames = glob.glob(raw_files)
track_uris = []

with concurrent.futures.ThreadPoolExecutor() as executor:
    dfs = list(executor.map(load_file, filenames))

df = pd.concat(dfs)
unique_tracks = np.unique(track_uris)

# Sanity check
assert df.shape[0] != 0 

#print the length of the track_uris
print(f'The length of the track_uris is: {len(unique_tracks)}')
print(f'The dataframe has {df.shape[0]} rows and {df.shape[1]} columns.')
````

## Fetching Metadata from SpotifyWEB API with Spotipy

Spotify provides a free web API to gain additional information about the songs, this will be incredibly useful for both PCA and K-means as the metadata includes things like:
- energy
- tempo

Having more generic about the data will allow us to infer overall characterictics of playlists.

NOTE: By default we will use the metadata from `metadata.csv`, to use the spotify API switch the `USE_SPOTIFY_API` variable to `True` 

````python
USE_SPOTIFY_API = False

#load spotify credentials from .env
client_id = os.getenv("SPOTIPY_CLIENT_ID")
client_secret = os.getenv("SPOTIPY_CLIENT_SECRET")
metadata_file = os.getenv("METADATA_FILE", "../docs/sample_data/metadata.csv")
batch_size = os.getenv("BATCH_SIZE", 100)

raw_features = []
metadata = []
processed_features = []
batches = [unique_tracks[x:x+batch_size] for x in range(0, len(unique_tracks), batch_size)]

#NOTE: the spotify api denies too many requests, take only the first 10000 tracks
batches = batches[:10]
unique_tracks = unique_tracks[:1000]

if client_id is None or client_secret is None or not USE_SPOTIFY_API:
    print("Using metadata from file.")
    # load metadata from metadata_file
    metadata = pd.read_csv(metadata_file)
else:
    print("Fetching metadata from Spotify.")
    #login to spotify
    sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=client_id, client_secret=client_secret))
    

    for batch in batches:
        batch_features = sp.audio_features(tracks = batch)
        raw_features.append(batch_features)

    raw_features = [item for sublist in raw_features for item in sublist]

    for track, features in zip(unique_tracks, raw_features):
                    processed_features.append({
                        "track_id":track,
                        "energy":features['energy'], 
                        "tempo":features['tempo'],
                        "speechiness":features['speechiness'],
                        "acousticness":features['acousticness'],
                        "instrumentalness":features['instrumentalness'],
                        "danceability":features['danceability'],
                        "loudness":features['loudness'],
                        "valence":features['valence'],
                        })

    metadata = pd.DataFrame(processed_features)
    # metadata.to_csv(metadata_file,sep=',',index=False)
````

## Unsupervised Methods
### Preprocessing metadata

We are going to be performing two types of visualisation, PCA and K-means clustering based visualisation. For both, the track IDs are irrelevant since we do not care about any particular track, but rather howthe tracks relate to each other. This means we can remove them. To aid the analysis, we will use the standard scaler which will normalise all variables to have `mean = 0` and `variance = 0`. This step will account for the natural difference in scales between the variables and produce more consistent and interesting results.

````python
# print(metadata.head())
metadata = metadata.drop(columns=["track_id"])
metadata = StandardScaler().fit_transform(metadata)
````

### PCA

Principal Component analysis (PCA) is a well documented dimension-reduction technique. We create a new _n-dimensional_ basis where _n_ corresponds to the number of principal components we are decomposing into. The basis is chosen such that when we project all of our data points onto this basis, the data preserves the largest amount of variance to the original data. 

Since we are using PCA to visualise data, the only choice of _n_ that makes sense is `n <= 3`. This will keep the final plot to be at most 3D which is the last dimension we can plot with relative ease. In theory, the choice of _n_ can be done with a similar "elbow method" as used later in K-means clustering.

After some experimentation, _n=2_ leads to the most interpretable results.

Interpreting PCA is arguably the most difficult, yet the most important part of this analysis. For the tracks, my best guess is that one of the components corresponds to the tempo of the song and the other corresponds to the overall mood of the song. This roughly matches up with the tracks present, but there is no way to verify this beyond a hypothesis.

````python
# Perform PCA on metadata 
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(metadata)
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
print(principalDf.head())

# plot the PCA
fig = plt.figure(figsize = (6,6))
ax = fig.add_subplot(1,1,1)
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
ax.scatter(principalDf['principal component 1'], principalDf['principal component 2'])
ax.grid()
plt.show()
````

### K-means

K-means clustering is used for grouping data points into distinct clusters.
The objective is to partition a dataset with n points into k clusters, where each point belongs to the cluster whose mean (centroid) is closest to it.
The algorithm aims to minimize the within-cluster sum of squares (WCSS) also known as inertia, which represents the variance within each cluster.
It optimizes squared errors, making it efficient for large datasets. 

### Finding k
The natural question is to ask: "How do we choose k?". The naive technique of minimizing the WCSS will not work in this case, since adding more clusters will always lead to be lower intertia while `k<n` and will go to zero with `k=n` where `n` is the total number of data points in the set, since each datapoint will become its own cluster, leading to zero inertia, but ruining any potential insights we might gain from the analysis.

A common method of choosing _k_ is the "elbow method", we plot inertia against our choice of _k_ then look for the point where inertia starts decreasing at a slower rate (forming an elbow). This point is where we get a good tradeoff between model complexity (number of clusters) and data fit. In our case we choose `k=4`.

````python
# try different values of k in a loop and plot the inertia
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(metadata)
    inertia.append(kmeans.inertia_)

# plot the different values of k and their inertia
plt.plot(range(1, 11), inertia)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
````

### K-means plot

Once we found _k_ we can perform the actual k-means clustering. We can use the principal components from our PCA analysis to visualise the results of the clusters.

````python
number_of_clusters = 4
kmeans = KMeans(n_clusters=number_of_clusters)
kmeans.fit(metadata)
y_kmeans = kmeans.predict(metadata)

plt.scatter(principalComponents[:, 0], principalComponents[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.show()
````
